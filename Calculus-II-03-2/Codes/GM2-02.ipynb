```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.linspace(0, 4*np.pi, 200)
x = np.sin(t)
y = np.cos(t)
```

```python
fig, ax = plt.subplots(figsize=(5,6), nrows=2, ncols=1)
ax[0].plot(t, x)
ax[1].plot(t, y)
plt.show()

```

```python
plt.figure(figsize=(3,3))
plt.plot(x,y)
plt.show()
```

```python
# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot(t, x, y)
plt.show()
```

```python
t = np.linspace(0, 10, 250)
x = t*np.exp(t)
y = np.cos(t)**2
z = 1/(t+1)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot(x, y, z)
plt.show()
```

```python
t = np.linspace(0, 10, 250)
x = np.sin(5*t)
y = np.cos(2*t+1)
z = np.sqrt(t)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot(x, y, z)
plt.show()
```
